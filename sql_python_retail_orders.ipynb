```python
import pandas as pd
df = pd.read_csv('orders.csv')
df.head(10) # gives 5 records by default, if not specified in the parenthesis
```

```python
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
# df.rename(columns={'Ship Mode':'ship_mode','Order Id':'order_id'})
# df.columns.str.lower()
df.columns = df.columns.str.lower()
# df.columns
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df.head()
```

```python
df['discount'] = df['list_price']*df['discount_percent']/100
```

```python
df.head()
```

```python
df['sale_price'] = df['list_price'] - df['discount']
```

```python
df.head()
```

```python
df['profit'] = (df['sale_price'] - df['cost_price']) * df['quantity']
```

```python
df.head()
```

```python
df['final_sale'] = df['sale_price'] * df['quantity']
```

```python
df.head()
```

```python
# Check the data types of all the columns

#df.dtypes

# the data type of oreder_date is object, need to convert it to datetime
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
df.dtypes
```

```python
# df.drop(columns = ['list_price','cost_price','discount_percent'])
# these columns are not actually dropped from the actual dataplace
# To check run 
# run df
# we need to add inlace = True
df.drop(columns = ['list_price','cost_price','discount_percent'], inplace = True)
# The columns are dropped now.
```
